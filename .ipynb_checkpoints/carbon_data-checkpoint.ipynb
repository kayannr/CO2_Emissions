```python
# Parsing of “students.xml” file starts at the root of the tree, namely the <data> element, 
# which contains the entire data structure.xtree = et.parse("students.xml")
xtree = et.parse("Data/worldbank.org(1960-2014)/API_EN.ATM.CO2E.KT_DS2_en_xml_v2_1073984.xml")
xroot = xtree.getroot() 
```

```python
df_cols = ["Country or Area", "Item", "Year", "Value"]
rows = []

for node in xroot: 
    s_country = node.attrib.get("Country or Area")
    s_item = node.find("Item").text if node is not None else None
    s_year = node.find("Year").text if node is not None else None
    s_value = node.find("Value").text if node is not None else None
    
    rows.append({"Country or Area": s_country, "Item": s_item, 
                 "Year": s_year, "Value": s_value})
```

```python
record = {"record":[]}
for elem in xroot:
    records = {}
    records["Country"] = elem.attrib['Country or Area']
    records["Item"] = elem.attrib['TagName']
    records["Year"] = elem.attrib['Count']
    records["Value"]= elem.attrib['Value']
    record["record"]. append(records)

df_users = pd.DataFrame(record["record"])
df_users.head()
```

```python
xml_data = "Data/worldbank.org(1960-2014)/API_EN.ATM.CO2E.KT_DS2_en_xml_v2_1073984.xml"
```

```python
class XML2DataFrame:

    def __init__(self, xtree):
        self.root = ET.XML(xtree)

    def parse_root(self, root):
        """Return a list of dictionaries from the text and attributes of the
        children under this XML root."""
        return [parse_element(child) for child in root.getchildren()]

    def parse_element(self, element, parsed=None):
        """ Collect {key:attribute} and {tag:text} from thie XML
         element and all its children into a single dictionary of strings."""
        if parsed is None:
            parsed = dict()

        for key in element.keys():
            if key not in parsed:
                parsed[key] = element.attrib.get(key)
            if element.text:
                parsed[element.tag] = element.text                
            else:
                raise ValueError('duplicate attribute {0} at element {1}'.format(key, element.getroottree().getpath(element)))           

        """ Apply recursion"""
        for child in list(element):
            self.parse_element(child, parsed)
        return parsed

    def process_data(self):
        """ Initiate the root XML, parse it, and return a dataframe"""
        structure_data = self.parse_root(self.root)
        return pd.DataFrame(structure_data)

xml2df = XML2DataFrame(xtree)
xml_dataframe = xml2df.process_data()
```
